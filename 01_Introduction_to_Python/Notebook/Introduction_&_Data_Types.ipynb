### 🐍 Introduction to Python and Why Python for ML?
Python is a high-level, general-purpose, and very popular programming language. <br>
It was created in 1991 by 'Guido van Rossum' and quickly became a widely used language<br>
in various fields, including web development, data science, machine learning, and automation

---

#### Key Features of Python:
-  High Readability:<br>Python is designed to maximize code readability.<br>It uses simple English for keywords and commands, and its code structure is clean and organized, making it easier to understand and maintain.
- Simple and Easy to Learn:<br>Python's syntax is simple and understandable, making it suitable for beginners with any background.
- Versatile:<br>Python can be used for a wide range of tasks, including web development, game development, data analysis, and more...
- Extensive Libraries:<br>Python has a very rich ecosystem of libraries and frameworks, meaning there's a solution available for practically anything you can imagine.
- Active Community and Strong Support:<br>Python has a large and active community of developers who are constantly creating new libraries, frameworks, and tools.

### Variable Naming Conventions
1. Variable names can contain only letters, numbers, and underscores.<br> 
They can start with a letter or an underscore, but not with a number.<br>
For instance, you can call a variable message_1 but not 1_message.

```python
message_1 = "hello"
1_message = "word"
```

2. Spaces are not allowed in variable names,but underscores can be used 
to separate words in variable names.<br>For example, greeting_message 
works, but greeting message will cause errors.


```python
greeting_message = "thanks"
greeting message = "hello"
```

4. Variable names should be short but descriptive .<br>
